```python
import pandas as pd
df=pd.read_csv("student-data.csv")

df.loc[0]
```

```python
filt=(df["sex"]=='M')&(df['romantic']=='yes')
df.loc[~filt,'age']
df.head()
```

```python
df["sex"] = df["sex"].str.strip().replace({
      "M": "Male",
          "F": "Female"
          })
```

```python
df.sort_values(by=['age','Medu'],ascending=[True,False])
```

```python
df.apply(pd.Series.min)
```

```python
fjgrp=df.groupby(('Fjob'))
fjgrp['absences'].agg(['median','mean']).loc['teacher']
```

```python
fjgrp=df.groupby(('Fjob'))
fjgrp.get_group('teacher')
```

```python
ts=fjgrp['Fjob'].value_counts()
ts
```

```python
f=df['Fjob']=='teacher'
df.loc[f]['passed'].str.contains('yes').sum()
```

```python
psfj=fjgrp['passed'].apply(lambda x:x.str.contains('yes').sum())
psfj
```

```python
dff=pd.concat([ts,psfj],axis='columns',sort=False)
dff['percentage']=(dff['passed']/dff['count'])*100
dff
```
